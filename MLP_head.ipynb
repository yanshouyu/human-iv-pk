In this notebook I'm going to build MLP as predictor on physiochemical properties and fingerprints. The MLP structure comes from `SmipleMLP` object in TAPE project.

```python
import numpy as np
import pandas as pd
import torch
from torch import nn
import torch.nn.functional as F
from torch.nn.utils.weight_norm import weight_norm
from tqdm import tqdm, trange
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%load_ext autoreload
%autoreload 2
import ivpk
```

```python
class SimpleRegHead(nn.Module):
    def __init__(self,
                 in_dim: int,
                 hid_dim: int, 
                 dropout: float = 0., 
                 intersept: float = None):
        super().__init__()
        self.main = nn.Sequential(
            nn.Linear(in_dim, hid_dim), 
            nn.GELU(), 
            nn.Dropout(dropout, inplace=False),
            nn.Linear(hid_dim, 1)
        )
        # set the bias if needed
        if intersept:
            with torch.no_grad():
                self.main[3].bias.fill_(intersept)

    def forward(self, x):
        return self.main(x)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

Use VDss as example target and physiochemical properties as example input.

```python
target = 'VDss'
```

Dataset and data loader. Be careful that the dimensionality for y should not be (N, ) but (N, 1) here. Otherwise the loss function will not calculate properly due to broadcasting rules.

```python
train_dl, val_dl = ivpk.data.get_dataloaders(target, ("train", "val"), smiles_func=None)
```

Dummy run training.

```python
reg = SimpleRegHead(11, 512).to(device)
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(reg.parameters(), lr=0.01)
```

```python
def loss_batch(model, loss_fn, xb, yb, opt=None):
    loss = loss_fn(model(xb), yb)

    if opt is not None:
        loss.backward()
        opt.step()
        opt.zero_grad()

    return loss.item(), len(xb)

def fit(num_epochs, model, loss_fn, optimizer, train_dl, val_dl):
    for epoch in trange(num_epochs):
        model.train()
        losses, nums = zip(
            *[loss_batch(model, loss_fn, Xb, yb, optimizer) for Xb, yb in train_dl]
        )
        train_loss = np.sum(np.multiply(losses, nums)) / np.sum(nums)
        model.eval()
        with torch.no_grad():
            losses, nums = zip(
                *[loss_batch(model, loss_fn, Xb, yb) for Xb, yb in val_dl]
            )
        val_loss = np.sum(np.multiply(losses, nums)) / np.sum(nums)
        yield (epoch, train_loss, val_loss)
```

```python
history = pd.DataFrame(
    fit(500, reg, loss_fn, optimizer, train_dl, val_dl), 
    columns=['epoch', 'train_loss', 'val_loss']
)
```

```python
plt.plot(history.epoch, history.train_loss)
plt.plot(history.epoch, history.val_loss)
plt.show()
```

```python
x_val, y_val = val_dl.dataset.tensors
plt.plot(y_val, reg(x_val).detach().numpy(), ".")
```

```python
x_train, y_train = train_dl.dataset.tensors
plt.plot(y_train, reg(x_train).detach().numpy(), ".")
```

```python
print(
    "MAE on val", 
    mean_absolute_error(y_val, reg(x_val).detach().numpy()), 
    "\nMAE on train", 
    mean_absolute_error(y_train, reg(x_train).detach().numpy())
)
```

## Conclusion

None of the MLP regressor can beat performance of even the default RFreg on properties (11) + morgan FP (256). 

- MAE \~1.1 (MSE 2.25\~2.50) is the limit for no-FP input in VDss (same for RFreg and MLP reg), as the train and val loss are close but still not low enough.
- High-variance in VDss on property + morgan256, thus regularization is needed.
- High-bias, high-variance in no-FP input for CL. We'll need **more information to support a proper prediction on CL**.
- High-variance in CL on property + morgan256, consider regularization.
