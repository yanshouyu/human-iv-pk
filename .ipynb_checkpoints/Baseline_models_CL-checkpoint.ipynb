This notebook is a mirror notebook for CL prediction baseline models.

Baseline strategy overview
- Baseline data
    - computed physiochemical properties (11) only
    - computed physiochemical properties (11) + Morgan FP (256)
    - computed physiochemical properties (11) + Morgan FP (2048)
- Baseline model
    - Linear regression
    - Random Forest Regressor
    - SVR
    - (Not implemented yet) Simple MLP (adopted from TAPE's regression head)
- Metrics

Note that here we're not going to enumerate all possible baseline combinations, but to explore a framework to try and capture results. Hyperparameter tuning will be another story. Here we explore which models should proceed to hyperparameter tuning.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%load_ext autoreload
%autoreload 2
import ivpk
```

```python
target = "CL"
```

## Plots

```python
def plot_model_eval(ev: ivpk.evalutaion.Evaluation):
    def plot_cl(x, y):
        plt.plot(x, y, ".")
        plt.xlim(-6, 8)
        plt.ylim(-6, 8)
        plt.xlabel("real")
        plt.ylabel("predicted")
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plot_cl(ev.y_val, ev.y_val_pred)
    plt.title((
        "validation", 
        f"MAE: {ev.evaluation['MAE_val']:.3f},", 
        f"Pearson: {ev.evaluation['Pearsonr_val']:.3f}"
    ))
    plt.subplot(122)
    plot_cl(ev.y_train_val, ev.y_train_val_pred)
    plt.title((
        f"{ev.evaluation['cv']}fold pred", 
        f"MAE: {ev.evaluation['MAE_cv']:.3f}", 
        f"Pearson: {ev.evaluation['Pearsonr_cv']:.3f}"
    ))
```

## Linear regression
Start from the simplest baseline, linear regression on computed physiochemical properties only.

```python
from sklearn.linear_model import LinearRegression
```

```python
eval_linreg_nofp = ivpk.evalutaion.Evaluation(
    LinearRegression(), target=target, validation_size=0.2, smiles_func=None)
```

```python
eval_linreg_nofp.evaluate(use_smiles=False, fit=True)
```

```python
plot_model_eval(eval_linreg_nofp)
```

Linear regression on pre-provided computed physiochemical properties cannot make a proper prediction. Let's add the fingerprints to X.

```python
eval_linreg_morgan256 = ivpk.evalutaion.Evaluation(
    LinearRegression(), target=target, validation_size=0.2, 
    fpType="morgan", fpSize=256
)

eval_linreg_morgan256.evaluate(use_smiles=False, fit=True)
```

```python
plot_model_eval(eval_linreg_morgan256)
```

Simple linear regression doesn't work w/ or wo/ fingerprints.

## Leaderboard

```python
leaderboard = pd.DataFrame(
    [
        (
            ev.evaluation["MAE_val"], 
            ev.evaluation["Pearsonr_val"], 
            ev.evaluation["MAE_cv"], 
            ev.evaluation["Pearsonr_cv"]
        ) for ev in (
            eval_linreg_nofp, eval_linreg_morgan256, 
            eval_lasso_nofp, eval_lasso_morgan256, 
            eval_rfreg_nofp, eval_rfreg_morgan256, 
            eval_svr_nofp, eval_svr_morgan256
        )
    ], 
    columns=["MAE_val", "Pearsonr_val", "MAE_cv", "Pearsonr_cv"], 
    index=pd.MultiIndex.from_product(
        [
            ["linreg", "lasso02", "RFreg", "SVR"],    # models
            ["properties", "properties+morgan256"],    # data
        ], 
        names=["model", "data"]
    )
)
```

## Model capacity

Let's see if the model can overfit on training set.

```python
ev_names = ["Lasso02", "RFreg", "SVR"]
plt.figure(figsize=(15, 5))
for i, ev in enumerate((
    eval_lasso_morgan256, 
    eval_rfreg_morgan256, 
    eval_svr_morgan256
)):
    plt.subplot(1, 3, i+1)
    plt.plot(
        ev.y_train, 
        ev.model.predict(ev.x_train[0]), 
        "."
    )
    plt.xlim(-4, 10)
    plt.ylim(-4, 10)
    plt.title(ev_names[i])
```

Surely RFreg can overfit its training set but the other 2 cannot. We can consider tune its hyperparameters on properties + morgan256. There is no need to explore any linear models.
