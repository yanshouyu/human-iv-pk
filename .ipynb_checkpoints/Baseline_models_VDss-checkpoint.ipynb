In this notebook we can start building some baseline models. In the meantime we can implement metrics and visualization for model evaluation. Since we have 2 targets, I'll focus on VDss in this notebook.

Baseline strategy overview
- Baseline data
    - computed physiochemical properties (11) only
    - computed physiochemical properties (11) + Morgan FP (256)
    - computed physiochemical properties (11) + Morgan FP (2048)
- Baseline model
    - Linear regression
    - Random Forest Regressor
    - SVR
    - Simple MLP (adopted from TAPE's regression head)
- Metrics
    - Pearson's correlation
    - Mean absolute error

Note that here we're not going to enumerate all possible baseline combinations, but to explore a framework to try and capture results. Hyperparameter tuning will be another story. Here we explore which models should proceed to hyperparameter tuning.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%load_ext autoreload
%autoreload 2
import ivpk
```

```python
from sklearn.metrics import mean_absolute_error
from scipy.stats import pearsonr
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
target = "VDss"
```

## Plots

```python
def plot_model_eval(ev: ivpk.evalutaion.Evaluation):
    def plot_vdss(x, y):
        plt.plot(x, y, ".")
        plt.xlim(-4, 6)
        plt.ylim(-4, 6)
        plt.xlabel("real")
        plt.ylabel("predicted")
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plot_vdss(ev.y_val, ev.y_val_pred)
    plt.title((
        "validation", 
        f"MAE: {ev.evaluation['MAE_val']:.3f},", 
        f"Pearson: {ev.evaluation['Pearsonr_val']:.3f}"
    ))
    plt.subplot(122)
    plot_vdss(ev.y_train_val, ev.y_train_val_pred)
    plt.title((
        f"{ev.evaluation['cv']}fold pred", 
        f"MAE: {ev.evaluation['MAE_cv']:.3f}", 
        f"Pearson: {ev.evaluation['Pearsonr_cv']:.3f}"
    ))
```

## Linear regression
Start from the simplest baseline, linear regression on computed physiochemical properties only.

```python
from sklearn.linear_model import LinearRegression
```

```python
eval_linreg_nofp = ivpk.evalutaion.Evaluation(
    LinearRegression(), target=target, validation_size=0.2, smiles_func=None)
```

```python
eval_linreg_nofp.evaluate(use_smiles=False, fit=True)
```

```python
plot_model_eval(eval_linreg_nofp)
```

Linear regression with pre-provided computed physiochemical properties can catch certain trend, but not satisfactory. Let's add the fingerprints to X.

```python
eval_linreg_morgan256 = ivpk.evalutaion.Evaluation(
    LinearRegression(), target=target, validation_size=0.2, 
    fpType="morgan", fpSize=256
)

eval_linreg_morgan256.evaluate(use_smiles=False, fit=True)
```

```python
plot_model_eval(eval_linreg_morgan256)
```

Adding a 256 morgan fingerprint doesn't make the metrics better but the predicted value span looks nicer.

```python
eval_lasso_morgan256.model.coef_
```

```python
eval_lasso_morgan1024 = ivpk.evalutaion.Evaluation(
    model=Lasso(alpha=0.02), target=target, validation_size=0.2, fpType="morgan", fpSize=1024
)
_ = eval_lasso_morgan1024.evaluate(use_smiles=False, fit=True, cv=5)
plot_model_eval(eval_lasso_morgan1024)
```

The model cannot follw extreme high values, there must be something in them not reflected by physiochemical values or fingerprints.

## Leaderboard

```python
leaderboard = pd.DataFrame(
    [
        (
            ev.evaluation["MAE_val"], 
            ev.evaluation["Pearsonr_val"], 
            ev.evaluation["MAE_cv"], 
            ev.evaluation["Pearsonr_cv"]
        ) for ev in (
            eval_linreg_nofp, eval_linreg_morgan256, 
            eval_lasso_nofp, eval_lasso_morgan256, 
            eval_rfreg_nofp, eval_rfreg_morgan256, 
            eval_svr_nofp, eval_svr_morgan256
        )
    ], 
    columns=["MAE_val", "Pearsonr_val", "MAE_cv", "Pearsonr_cv"], 
    index=pd.MultiIndex.from_product(
        [
            ["linreg", "lasso02", "RFreg", "SVR"],    # models
            ["properties", "properties+morgan256"],    # data
        ], 
        names=["model", "data"]
    )
)
```

## Model capacity

Let's see if the model can overfit on training set.

```python
ev_names = ["Lasso02", "RFreg", "SVR"]
plt.figure(figsize=(15, 5))
for i, ev in enumerate((
    eval_lasso_morgan256, 
    eval_rfreg_morgan256, 
    eval_svr_morgan256
)):
    plt.subplot(1, 3, i+1)
    plt.plot(
        ev.y_train, 
        ev.model.predict(ev.x_train[0]), 
        "."
    )
    plt.xlim(-4, 10)
    plt.ylim(-4, 10)
    plt.title(ev_names[i])
```

Surely RFreg can overfit its training set but the other 2 cannot. We can consider tune its hyperparameters on properties + morgan256. For LASSO regression we can also consider tune its hyperparameters to push the performance of a linear baseline.

We don't consider L2 regularization since we want to squeeze the input dimensionality.
